```python
### cmc acts
from sklearn.metrics import classification_report
import pandas as pd
true=df[['cmc_acts_true']]
pred=predictions[['annot_act']].apply(lambda x: x.str.lower().str.strip())
assert len(true)==len(pred), "Length mismatch between true labels and predictions"
tmp=pd.concat([true.reset_index(drop=True), pred.reset_index(drop=True)], axis=1)
tmp.columns=['true_label', 'predicted_label']
# Expand multi-labels into separate rows
expanded_rows = []
for idx, row in tmp.iterrows():
    valid_labels = [label.strip() for label in row['true_label'].split('/')]
    for label in valid_labels:
        expanded_rows.append({
            'true_label': label,
            'predicted_label': row['predicted_label'],
            'original_idx': idx
        })

expanded_df = pd.DataFrame(expanded_rows)

# Group by original index and check if ANY match
results = expanded_df.groupby('original_idx').apply(
    lambda x: (x['true_label'] == x['predicted_label']).any()
)

def check_correct(true_label, predicted_label):
    """Check if prediction matches any valid label"""
    valid_labels = [label.strip() for label in true_label.split('/')]
    return predicted_label in valid_labels

# Apply to your dataframe
tmp['correct'] = tmp.apply(lambda row: check_correct(row['true_label'], row['predicted_label']), axis=1)

# Calculate accuracy
accuracy = tmp['correct'].mean()
print(f'Accuracy: {accuracy:.4f}')

# For detailed metrics per class (use first valid label as ground truth)
tmp['primary_label'] = tmp['true_label'].str.split('/').str[0].str.strip()
print(classification_report(tmp['primary_label'], tmp['predicted_label']))
```

```python
from sklearn.metrics import cohen_kappa_score
import numpy as np

def prepare_for_kappa(df, rater1_col, rater2_col):
    """
    Prepare multi-label data for Cohen's Kappa
    Treats any matching label as agreement
    """
    agreements = []
    
    for _, row in df.iterrows():
        rater1_labels = set([l.strip() for l in str(row[rater1_col]).split('/')])
        rater2_label = str(row[rater2_col]).strip()
        
        # Check if rater2's label matches any of rater1's valid labels
        if rater2_label in rater1_labels:
            # Agreement - both get the same label
            agreements.append((rater2_label, rater2_label))
        else:
            # Disagreement - use first valid label for rater1
            agreements.append((list(rater1_labels)[0], rater2_label))
    
    y1, y2 = zip(*agreements)
    return list(y1), list(y2)

# Calculate Cohen's Kappa
y_true, y_pred = prepare_for_kappa(tmp, 'true_label', 'predicted_label')
kappa = cohen_kappa_score(y_true, y_pred)
print(f"Cohen's Kappa: {kappa:.3f}")
```

```python
#### politeness
from sklearn.metrics import classification_report
import pandas as pd
true=df[['politeness_true']]
pred=predictions[['annot_politness_new']].apply(lambda x: x.str.strip())
assert len(true)==len(pred), "Length mismatch between true labels and predictions"
tmp=pd.concat([true.reset_index(drop=True), pred.reset_index(drop=True)], axis=1)
tmp.columns=['true_label', 'predicted_label']
tmp['true_label'] = tmp['true_label'].fillna('Unknown').astype(str)
tmp['predicted_label'] = tmp['predicted_label'].fillna('Unknown').astype(str)
# Calculate accuracy for all rows
accuracy = (tmp['true_label'] == tmp['predicted_label']).mean()
print(f'Accuracy: {accuracy:.4f}')
print(classification_report(tmp['true_label'], tmp['predicted_label']))

# calculate cohen's kappa
y_true, y_pred = prepare_for_kappa(tmp, 'true_label', 'predicted_label')
kappa = cohen_kappa_score(y_true, y_pred)
print(f"Cohen's Kappa: {kappa:.3f}")

# calculate accuracy excluding 'Unknown'
filtered_tmp = tmp[tmp['true_label'] != 'Unknown']
accuracy_filtered = (filtered_tmp['true_label'] == filtered_tmp['predicted_label']).mean()
print(f'Accuracy (excluding Unknown): {accuracy_filtered:.4f}') 
print(classification_report(filtered_tmp['true_label'], filtered_tmp['predicted_label']))
```
